# Heat Pump Indicator Data Wrangling 

```python
import os
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt 
import numpy as np
```

```python
os.getcwd() 
```

```python
wkdir = os.getcwd()
accommodation = pd.read_csv(wkdir + '/datasets/accommodation_type_census_2021.csv')
heating = pd.read_csv(wkdir + '/datasets/central_heating_uk_2021_census.csv')
income = pd.read_excel(wkdir + '/datasets/la_district_imd_scores.xlsx', sheet_name='Income')
employment = pd.read_excel(wkdir + '/datasets/la_district_imd_scores.xlsx', sheet_name='Employment')
tenure = pd.read_csv(wkdir + '/datasets/tenure_census_2021.csv')
insulation = pd.read_excel(wkdir + '/datasets/INSULATION_ANALYSIS_ALL_AREA_DATA_0.xlsx', sheet_name='Local authority') 
fuel_poverty = pd.read_excel(wkdir + '/datasets/sub_regional_fuel_poverty_2021.xlsx', sheet_name='Table 2', skiprows=2) 
```

```python
accommodation = accommodation.rename(columns={'Lower tier local authorities Code':'Area Code', 'Accommodation type (5 categories) Code':'Type Code'})
accommodation['Household No.s'] =  accommodation['Observation'].groupby(accommodation['Area Code']).transform('sum')
accommodation
```

```python
all_house = accommodation[(accommodation['Type Code']>=1) & (accommodation['Type Code']<=3)]
all_house
```

```python
all_house['All Houses'] = all_house['Observation'].groupby(all_house['Area Code']).transform('sum')
```

```python
all_house
```

```python
all_house['All Houses (%)'] = all_house['All Houses']/all_house['Household No.s']*100
all_house = all_house.drop_duplicates(subset=['Area Code'])
all_houses = all_house[['Area Code', 'All Houses', 'All Houses (%)']]
all_houses
```

```python
all_house_detach_semi = accommodation[(accommodation['Type Code']==1) | (accommodation['Type Code']==2)]
all_house_detach_semi
```

```python
all_house_detach_semi['All Houses - Detached or Semi'] = all_house_detach_semi['Observation'].groupby(all_house_detach_semi['Area Code']).transform('sum')
all_house_detach_semi
```

```python
all_house_detach_semi['All Houses - Detached or Semi (%)'] = all_house_detach_semi['All Houses - Detached or Semi']/all_house_detach_semi['Household No.s']*100
all_house_detach_semi = all_house_detach_semi.drop_duplicates(subset=['Area Code'])
all_houses_detach_semi = all_house_detach_semi[['Area Code', 'All Houses - Detached or Semi', 'All Houses - Detached or Semi (%)']]
all_houses_detach_semi
```

```python
non_house = accommodation[(accommodation['Type Code']==4) | (accommodation['Type Code']==5)]
non_house
```

```python
non_house['Non-House Accommodation'] = non_house['Observation'].groupby(non_house['Area Code']).transform('sum')
non_house['Non-House Accommodation (%)'] = non_house['Non-House Accommodation']/non_house['Household No.s']*100
non_house = non_house[['Area Code', 'Non-House Accommodation', 'Non-House Accommodation (%)']]
non_house
```

```python
housing = pd.merge(all_houses, all_houses_detach_semi, on='Area Code')
```

```python
housing = pd.merge(housing, non_house, on='Area Code')
housing
```

```python
heating = heating.rename(columns={'Lower tier local authorities Code':'Area Code', 'Lower tier local authorities':'Name', 'Type of central heating in household (13 categories) Code':'Heating Type Code', 'Type of central heating in household (13 categories)':'Central Heating Type'})
heating['Household No.s'] = heating['Observation'].groupby(heating['Area Code']).transform('sum') #household no.s from separate data
heating['Observation (%)'] = heating['Observation']/heating['Household No.s']*100
heating
```

```python
mains_gas_heating = heating[heating['Central Heating Type'] == 'Mains gas only'].rename(columns={'Observation':'Mains gas only', 'Observation (%)':'Mains gas only (%)'})
mains_gas_heating = mains_gas_heating[['Area Code','Name','Mains gas only','Mains gas only (%)']]

tank_bottle_gas = heating[heating['Central Heating Type'] == 'Tank or bottled gas only'].rename(columns={'Observation':'Tank or bottled gas only', 'Observation (%)':'Tank or bottled gas only (%)'})
tank_bottle_gas = tank_bottle_gas[['Area Code','Name','Tank or bottled gas only','Tank or bottled gas only (%)']]
```

```python
mains_gas_heating
```

```python
tank_bottle_gas
```

```python
gas_heating = heating[(heating['Central Heating Type'] == 'Mains gas only') | (heating['Central Heating Type'] == 'Tank or bottled gas only')]
gas_heating
```

```python
gas_heating['Gas only'] = gas_heating['Observation'].groupby(gas_heating['Area Code']).transform('sum')
```

```python
gas_heating = gas_heating.drop_duplicates(subset=['Area Code'])
```

```python
gas_heating['Gas only (%)'] = gas_heating['Gas only']/gas_heating['Household No.s']*100
gas_heating = gas_heating[['Area Code', 'Name', 'Gas only', 'Gas only (%)']]
gas_heating
```

```python
electric_heating = heating[heating['Central Heating Type'] == 'Electric only'].rename(columns={'Observation':'Electric only', 'Observation (%)':'Electric only (%)'})
electric_heating = electric_heating[['Area Code','Name', 'Electric only', 'Electric only (%)']]

oil_heating = heating[heating['Central Heating Type'] == 'Oil only'].rename(columns={'Observation':'Oil only', 'Observation (%)':'Oil only (%)'})
```

```python
electric_heating
```

```python
oil_heating = oil_heating[['Area Code', 'Name', 'Oil only', 'Oil only (%)']]
oil_heating
```

```python
all_heating = pd.merge(mains_gas_heating, tank_bottle_gas, on='Area Code')
all_heating
```

```python
all_heating = pd.merge(all_heating, gas_heating, on='Area Code')
```

```python
all_heating = pd.merge(all_heating, electric_heating.drop(['Name'], axis=1), on='Area Code')
all_heating = pd.merge(all_heating, oil_heating.drop(['Name'], axis=1), on='Area Code')
all_heating
```

```python
all_heating = all_heating.drop(['Name_x', 'Name_y'], axis=1)
```

```python
all_heating
```

```python
tenure = tenure.rename({'Lower tier local authorities Code':'Area Code', 'Lower tier local authorities':'Name', 'Tenure of household (7 categories) Code':'Tenure Code', 'Tenure of household (7 categories)':'Tenure'}, axis=1)
tenure['Household No.s'] = tenure['Observation'].groupby(tenure['Area Code']).transform('sum')
tenure['Observation (%)'] = tenure['Observation'] / tenure['Household No.s'] * 100
tenure
```

```python
tenure
```

```python
tenure_owned = tenure[(tenure['Tenure Code']==0) | (tenure['Tenure Code']==1)]
tenure_owned['Tenure - Owned'] = tenure_owned['Observation'].groupby(tenure_owned['Area Code']).transform('sum')
tenure_owned['Tenure - Owned (%)'] = tenure_owned['Tenure - Owned']/tenure_owned['Household No.s']*100
tenure_owned = tenure_owned.drop_duplicates(subset=['Area Code'])
tenure_owned = tenure_owned[['Area Code','Tenure - Owned', 'Tenure - Owned (%)']]
tenure_owned
```

```python
tenure_social = tenure[(tenure['Tenure Code']==2) | (tenure['Tenure Code']==3)]
tenure_social['Tenure - Social'] = tenure_social['Observation'].groupby(tenure_social['Area Code']).transform('sum')
tenure_social['Tenure - Social (%)'] = tenure_social['Tenure - Social']/tenure_social['Household No.s']*100
tenure_social = tenure_social.drop_duplicates(subset=['Area Code'])
tenure_social = tenure_social[['Area Code','Tenure - Social', 'Tenure - Social (%)']]
tenure_social
```

```python
tenure_private = tenure[(tenure['Tenure Code']==4) | (tenure['Tenure Code']==4)]
tenure_private['Tenure - Private Rent'] = tenure_private['Observation'].groupby(tenure_private['Area Code']).transform('sum')
tenure_private['Tenure - Private Rent (%)'] = tenure_private['Tenure - Private Rent']/tenure_private['Household No.s']*100
tenure_private = tenure_private.drop_duplicates(subset=['Area Code'])
tenure_private = tenure_private[['Area Code','Tenure - Private Rent', 'Tenure - Private Rent (%)']]
tenure_private
```

```python
all_tenure = pd.merge(tenure_owned, tenure_social, on='Area Code')
all_tenure = pd.merge(all_tenure, tenure_private, on='Area Code')
all_tenure
```

```python
income = income.rename(columns={'Local Authority District code (2019)':'Area Code', 'Local Authority District name (2019)':'Name'})
income = income[['Area Code', 'Name', 'Income - Average score ']]
income
```

```python
employment = employment.rename(columns={'Local Authority District code (2019)':'Area Code', 'Local Authority District name (2019)':'Name'})
employment = employment[['Area Code', 'Name', 'Employment - Average score ']]
employment
```

```python
deprivation_scores = pd.merge(income, employment.drop(['Name'], axis=1), on='Area Code')
deprivation_scores
```

```python
imd = pd.read_excel(wkdir + '/datasets/imd.xlsx', sheet_name='IMD2019')
imd['Average Index of Multiple Deprivation (IMD) Decile'] = imd['Index of Multiple Deprivation (IMD) Decile'].groupby(imd['Local Authority District code (2019)']).transform(np.mean)
imd['Max Index of Multiple Deprivation (IMD) Decile'] = imd['Index of Multiple Deprivation (IMD) Decile'].groupby(imd['Local Authority District code (2019)']).transform('min')
imd = imd[['Local Authority District code (2019)', 'Local Authority District name (2019)', 'Average Index of Multiple Deprivation (IMD) Decile', 'Max Index of Multiple Deprivation (IMD) Decile']]
```

```python
imd = imd.rename(columns={'Local Authority District code (2019)':'Area Code', 'Local Authority District name (2019)':'Name'})
imd
```

```python
imd_grouped = imd.drop_duplicates()
imd_grouped
```

```python
deprivation = pd.merge(deprivation_scores, imd_grouped.drop(['Name'], axis=1), on='Area Code')
deprivation
```

```python
fuel_poverty = fuel_poverty.rename(columns={'Area Codes [Note 4]':'Area Code','Area names ':'Name'})
fuel_poverty = fuel_poverty[['Area Code', 'Name', 'Number of households in fuel poverty', 'Proportion of households fuel poor (%)']]
fuel_poverty
```

```python
insulation
```

```python
insulation = insulation.rename(columns={'LA Code':'Area Code'})
insulation = insulation[['Area Code', 'Proportion of homes needing loft insulation', 'Proportion of homes needing cavity wall insulation', 'Number of homes needing loft insulation', 'Number of homes needing cavity wall insulation']]
insulation
```

```python
housing = housing.drop_duplicates(subset=['Area Code'])
housing
```

```python
all_heating = all_heating.drop(['Name'], axis=1)
all_heating
```

```python
all_data = pd.merge(deprivation, fuel_poverty, on='Area Code')
all_data = pd.merge(all_data, all_heating, on='Area Code')
all_data = pd.merge(all_data, insulation, on='Area Code')
all_data = pd.merge(all_data, housing, on='Area Code')
all_data = pd.merge(all_data, all_tenure, on='Area Code')
all_data
```

```python
all_data = all_data.drop(['Name_y'], axis=1).rename(columns={'Name_x':'LA Name'})
all_data
```

```python
all_data.to_excel('heat_pump_indicator_data.xlsx')
all_data.to_csv('heat_pump_indicator_data.csv')
```
